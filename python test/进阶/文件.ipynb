# 文件操作
## 文件的编码
- 文件通过编码技术将内容翻译成二进制数从而被保存
- 通过正确的编码方式才能得到正确的编码
- 一般用UTF-8,为当下的通用编码，除此之外还有GBK,Big5

## 文件的读写操作
- 文件操作包括*打开、关闭、读、写*
- 操作步骤：打开，读写，关闭
### 打开操作
open(name,mode,encoding)
name为文件名,可指定路径；mode为打开文件的模式：只读(r)，写入(w)，追加(a)等；encoding为编码格式
```python
f = open('python.txt','r',encoding ="UTF-8")
```
### 读操作
- 文件对象.read(num): num表示从文件中读取的数据的长度，若无num则读取所有数据，得到字符串
- 文件对象.readlines()：可以将整个文件的内容一次性读取，并且返回一个列表
- 文件对象.readline()：只读一行数据
- 还可以用for循环来得到每一行的数据
- 文件对象.close()：关闭文件，不关闭那这个文件将一直被python程序占用
- with open("python.txt","r")as f:
    f.readlines()可以自动关闭文件，避免遗忘close

````python
#打开文件
f = open('D:/python.txt','r',encoding='utf-8')
print(type(f))
#读取操作
print(f'读取十字节结果：{f.read(10)}')
print(f'读取十字节结果：{f.read()}')    #连续调用时会从上个read后结尾处继续读取
# readlines
lines = f.readlines()   #读取文件的所有行，封装到列表中
print(type(lines))  #读取文件后无论什么方法都从上一个读取的结尾处继续读写
# readline
line1 = f.readline()
print(type(line1))
print(line1)
# for
for line in f:
    print(line)
# close
f.close()
# 自动关闭
with open("D:/python.txt",'r',encoding='utf-8') as f:
    for line in f:
        print(line)
````

## 文件的写入
```python
f= open("python.txt",'w')   #打开文件
f.write("hello world")  #文件写入
f.flush()   #内容刷新
```
- 直接调用write，内容是攒在缓冲区中而非真正写入文件
- 调用flush才能真正写入
- 该操作是为了避免频繁操作

````python
import time
#文件不存在
f = open("D:/text.txt",'w',encoding='utf-8')    #文件不存在，则自动创造
f.write("hello world")
f.flush()
f.close()   #close内置flush操作
#文件存在
f = open("D:/text.txt",'w',encoding='utf-8')
f.write("max min")
f.close()   #文件存在的话会删除内部内容从而重新写入
````

## 文件的追加
- 只需要将w替换为a，其余步骤一样

````python
f = open("D:/text.txt",'a',encoding='utf-8')
f.write("\nmax min")
f.close()
````
