# 面向对象编程
## 类与对象
- 类：所有具有相同*属性*和*功能*的事物称作一个类
- 对象：具有相同*属性*和*功能*的具体事物
- 可用类与对象将现实的关系转化为代码结构

````python
import json
import random

from _ctypes_test import func


#以狗为例子
class Dog:
    leg_num = 4 #属性
    has_hair = True
    has_tail = True

    def bark(self): #方法
        print("狂吠")

    def bite(self):
        print("咬人")

#对象实例化
alex = Dog()    #拥有dog的所有属性和方法
print(alex.leg_num)
alex.bark()
````

- 类中定义成员方法与函数有一定不同
```python
def 方法名(self, 形参1, ...):
    方法体
```
- self的存在是用于在类内部用成员方法调用成员变量
- self虽然在参数列表中，但是传参时可以忽略它，有其他形参时需要传参，没有时不用传参

- 类相当于设计图纸，对象相当于实体，只有基于类创建对象，才能进行具体的工作；

#### 小于符号__lt__
- 直接比较两个类对象会报错

````python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

stu1 = Student('dasvi',11)
stu2 = Student('dania',12)
print(stu1<stu2)
````

- 但在类内部实现__lt__，可同时完成小于和大于

````python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __lt__(self, other):
        return self.age < other.age

stu1 = Student('dasvi',11)
stu2 = Student('dania',12)
print(stu1<stu2) #Ture
print(stu1>stu2)    #False
````

#### 小于等于与大于等于__le__
- 具体的使用和小于一样
#### 等于__eq__
- 同上

````python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __le__(self, other):
        return self.age <= other.age

    def __eq__(self, other):
        return self.age == other.age

stu1 = Student('dasvi',11)
stu2 = Student('dania',12)
stu3 = Student('david',11)
print(stu1<=stu2)    #Ture
print(stu1>=stu2)    #False
print(stu1 == stu3)
````

## 封装
### 封装的概念
- 面向对象的三大特征：封装、继承、多态
- 封装就是将现实的事物描述为程序中的类
### 私有行为
- 现实中有私有的行为，故类中也应该有对应的形式
- 私有成员的定义：
    - 私有成员变量：__变量名
    - 私有成员方法：__方法名
```python
class Phone:
    imei = None #序列号
    producer = None

    __current_voltage = None #当前电压

    def call_by_5g(self):
        print('5g已开')

    def __keep_single_core(self):
        print('cpu以单核运行')
```
- 调用私有对象的变量和方法均报错，对私有对象赋值虽然不报错，但是该操作无效

````python
class Phone:
    imei = None #序列号
    producer = None

    __current_voltage = None #当前电压

    def call_by_5g(self):
        print('5g已开')

    def __keep_single_core(self):
        print('cpu以单核运行')

phone1 = Phone()
phone1.__keep_single_core()
````

- 私有成员虽然类对象不能用，但是可以被其他成员调用

````python
class Phone:
    __current_voltage = 0.5 #当前电压

    def __keep_single_core(self):
        print('cpu以单核运行')

    def call_by_5g(self):
        if self.__current_voltage >= 1:
            print('5g已开')
        else:
            print('电量不足')
phone2 = Phone()
phone2.call_by_5g()
````

- 私有成员存在的意义就是提供内部其他成员使用，不提供给类对象使用

## 继承
### 概念
- 每次更新不可能完全重新重写一个类，所以一般会在已有的类的基础上进行修改，因而有了继承
### 单继承
```python
class Phone:
    imei = None #序列号
    producer = None

    def call_by_4g(self):
        print('4g已开')

class Phone2026(Phone):
    face_id = True

    def call_by_5g(self):
        print('5g已开')
```
- 语法：class 类名(父类):
        内容体
- 子类能完全使用父类的所有语法（但包含私有成员）
- 继承分为：单继承、多继承

````python
class Phone:
    imei = None #序列号
    producer = None

    def call_by_4g(self):
        print('4g已开')

class Phone2026(Phone):
    face_id = True

    def call_by_5g(self):
        print('5g已开')

phone3 = Phone2026()
phone3.call_by_4g()
print(phone3.face_id)
````

### 多继承
- 一个子类继承多个父类
- 语法：class 类名(父类1，父类2，....):
            内容体
- 如果没有补充，则类主体写pass,用来补全语法
- 注意事项：多个父类中如果有同名的成员，则默认以继承顺序（即谁先来的优先级高）作为优先级
  即：先继承的保留，后继承的被覆盖

### 复写父类成员
- 即在子类中重新定义同名的属性或方法
```python
class phone:
    producer = 'IT'

    def call_by_5g(self):
        print('父类5g通话')

class phone2(phone):
    producer = 'II'

    def call_by_5g(self):
        print('子类5g通话')
```


````python
class phone:
    producer = 'IT'

    def call_by_5g(self):
        print('父类5g通话')

class phone2(phone):
    producer = 'II'

    def call_by_5g(self):
        print('子类5g通话')

phone4 = phone2()
phone4.call_by_5g()
````

- 如果调用被复写父类成员，则需要特殊的调用(仅可以在子类内部调用父类同名成员)
- 方法一：
    - 调用成员变量：父类名.成员变量
    - 调用成员方法: 父类名.成员方法(self)
- 方法二：
    - 调用成员变量：super().成员变量
    - 调用成员方法: super().成员方法()

````python
class phone:
    imei = None
    producer = 'IT'

    def call_by_5g(self):
        print('父类5g通话')

class myphone(phone):
    producer = "II"

    def call_by_5g(self):
        print(f'父类厂商为{phone.producer}')
        phone.call_by_5g(self)

phone4 = myphone()
phone4.call_by_5g()
````

````python
class phone:
    imei = None
    producer = 'IT'

    def call_by_5g(self):
        print('父类5g通话')

class myphone(phone):
    producer = "II"

    def call_by_5g(self):
        print(f'父类厂商为{super().producer}')
        super().call_by_5g()

phone4 = myphone()
phone4.call_by_5g()
````

## 类型注解
### 变量的类型注解
- 为变量设置类型注解
- 语法：变量:类型
- 适用于基础数据类型和类对象类型，以及数据容器类型

````python
# 基础数据类型注解
var_1 : int = 10
var_2 : str = 'dasvi'
# 类对象注解
class bird:
    pass
bird1 : bird = bird()
# 基础容器注解
my_list : list = [1,2,3]
my_tuple : tuple = (1,2,3)
my_dict : dict = {'dasvi':12}
# 容器详细注解
my_list : list[int] = [1,2,3]
my_tuple : tuple[int, str, bool] = (1,'dania',True)
my_dict : dict[str,int] = {'dasvi':12}
````

- 此外，还可以在注释中进行注解
- 语法：type:类型

````python
import random

class bird:
    pass

# alt + 回车自动导入包
var_1 = random.randint(1,10)    # type: int
def func():
    pass
var_3 = func()  # type: bird
````

- 类型注解仅仅只是提示作用，并非决定作用，即使数据类型和注解类型不对应也不会导致错误

## 多态
### 概念
- 多种状态，即完成某个行为时，使用不同的对象会得到不同的状态
- 对于同一个行为，传入不同的对象，得到不同的行为
- 多态常用于继承关系上，函数形参声明接受父类对象，实际传入子类对象进行工作

````python
class animal:
    def speak(self):
        pass

class Dog(animal):
    def speak(self):
        print('www')

class Cat(animal):
    def speak(self):
        print('mmm')

def noise(animal: animal):
    animal.speak()

dog = Dog()
cat = Cat()

noise(dog)
noise(cat)
````

- 不难看出，父类的方法是空实现，这样设计是：
    - 父类决定有哪些方法
    - 具体怎么实现由子类自行决定
- 这种写法称为抽象类（接口）
- 方法体为空实现的为抽象方法
- 抽象类就相当于定义了一个标准，由子类实现抽象方法
