## 列表的遍历
### while与for
如何遍历列表的元素？——while循环

如何取出元素？——列表[下标]

如何控制？——从0开始，列表下标<列表的元素数量

````python
#eg
def list_while_funct():
    my_list1 = [100, 200, 300]
    index = 0
    while index < len(my_list1):
        print(my_list1[index])
        index += 1
````

接下来是for循环

语法为：

for 临时变量 in 数据容器:

    对临时变量进行处理

````python
def list_for_funct():
    my_list2 = [111, 222, 333]
    for element in my_list2:
        print(element)
````

````python
list_for_funct()
list_while_funct()
````

#### 二者之间的不同
- 循环控制上
    - while可以自定义循环条件
    - for不可以，只能一个个从容器中取出
- 无限循环上：while可以，for不行
- 使用场景上：
    - while使用的范围大
    - for只适合较为简单的遍历

## 元组
### 元组的用处
与列表最大不同点是定义完成后不可修改，相当于只读list

### 元组的定义
- 使用小括号（），用逗号隔开各个数据，数据可以为不同的数据类型
- (元素,元素,...,元素)——元组字面量定义
- 变量 = (元素,元素,...,元素)——元组变量定义
- 变量 = （）/变量 = tuple()——定义空元组
    - 注意，定义一个元素其元素后要加逗号
- 元组也可以嵌套：（（），（））,也可以用索引取数

````python
(1, "hello", True)
t1 = (1, "hello", True)
t2 = tuple()
t3 = ()
print(type(t1))
print(type(t2))
print(type(t3))

t4 = ("hello")
t5 = ('hello',)
print(type(t4))
print(type(t5))

t6 = ((1,),(1,2))
print(type(t6))

print(t6[1][1])
````

### 元组的功能
- index:查找数据，如果有则输出下标
- count：统计某个元素在元组中出现的次数
- len()：统计元组元素个数

````python
t7 = (1, 2, 3, 4, 5, 6, 7)
print(t7.index(2))
print(t7.count(2))
print(len(t7))
````

### 元组的遍历——同列表

````python
#while
index = 0
while index < len(t7):
    print(t7[index])
    index += 1

#for
for element in t7:
    print(element)
````

### 注意事项
虽然元组不可修改，但元组内若有列表则可以修改

````python
t8 = (1, 3, [3, 4, 5])
t8[2][1] = 6
print(t8)
````

## 字符串
### 字符串定义
- 字符串是字符的容器，可以放任意数量的字符
- 支持下标索引，和其他容器一样，从前向后从0开始，从后向前从-1开始
    - 同元组，字符串也不可修改，如若修改只能得到新字符串

````python
# 下标索引
my_str = "my string"
print(type(my_str))
print(my_str[0])
print(my_str[-2])
````

### 字符串的常见操作
- index：查找特定字符串的下标索引值
- replace: 字符串替换，通过将字符串内特定的所有字符串换成另外的字符串
    - 不改变本身，而是得到新的字符串
- split:字符串的分割，按照指定的分割字符串，将字符串分为多个字符串并存入到列表对象中
    - 字符串本身不变，得到新的列表对象
- strip: 字符串的规整操作，其中strip()为取出前后空格，strip(字符串)则是取出指定字符串，去除字符串的内容不论顺序，只要有就直接删除
- 统计字符串中某字符串出现的次数以及统计字符串长度同上：count，len
- 遍历也同上

````python
my_str = 'itcast and itheima'
#index
index = my_str.index('and')
print(index)    #结果7
#replace
new = my_str.replace('itheima', 'itcast')
print(new)
print(my_str)
#split
my_str_list = my_str.split(' ')
print(f"将字符串{my_str}分割成{my_str_list}，该类型为{type(my_str_list)}")
my_str_list = my_str.split('and')
print(f"将字符串{my_str}分割成{my_str_list}，该类型为{type(my_str_list)}")  #去除指定的字符串
#strip
my_str = '  itcast and itheima  '
new_my_str = my_str.strip()
print(new_my_str)
my_str = '12itcast and itheima21'
new_my_str = my_str.strip('12ia')
print(new_my_str)
#count && len
my_str = 'itcast and itheima'
print(my_str.count('and'))
print(len(my_str))
````

## 序列与切片操作
### 序列
- 定义：内容连续，有序且可使用下标索引的数据容器
    - 例如：列表，元组，字符串

### 序列的常用操作
- 切片操作：从一个序列中取出一个子序列
    - 语法：序列[初始下标：结束下标：步长]
        - 初始下标留空表示从头开始(结果包含)
        - 结束下标留空表示到最后结束(结果不包含)
        - 步长表示取元素的间隔：1为一个个取数(默认，可不写)，2为隔一个数取数，N表示跳N-1个数取数，步长为负数则反向取数，但首尾下标要反向标志
    - 该操作不影响序列本身

````python
my_list = [0,1,2,3,4,5,6]
result1 = my_list[1:5]
print(result1)
my_tuple = (0,1,2,3,4,5,6)
result2 = my_tuple[ : :3]
print(result2)
my_str = '123456'
result3 = my_str[5:1:-1]
print(result3)
print(my_str[::-1]) #倒序输出
````

## 集合
### 集合的定义
- 集合的特点就是不重复，可去重且无序
- 基本语法：
```python
# 定义集合字面量
{元素，元素，......，元素}
# 定义集合变量
变量 = {元素，元素，......，元素}
# 定义空集合
变量 = set()
```


### 集合的基本操作
- 首先集合无序，所以集合不支持下标索引访问，但集合支持修改
- add：添加元素，集合本身被修改（只有集合中不存在的元素才修改）
- remove：去除元素
- pop：在集合中随机去除一个元素
- clear：清空集合
- 集合1.difference(集合2)：取出集合1和集合2的差集，得到新集合但集合1，2不变
- 集合1.difference_update(集合2)：删除集合1中和集合二相同的元素，集合1修改，集合2不变
- 集合1.union(集合2)：将集合1和集合2组合成新集合，集合1，2不变
- len：统计集合元素个数
- 集合的遍历：由于不能用下标，所以不可用while，只能用for

````python
my_set = {"add","sub","multi","sum"}
# add
my_set.add("hello")
print(my_set)
# remove
my_set.remove("hello")
print(my_set)
# pop
element = my_set.pop()
print(element)
print(my_set)
#clear
my_set.clear()
print(my_set)
#difference
set1 = {1,2,3,4}
set2 = {2,3,5,6}
set3 = set1.difference(set2)
print(set3)
#difference_update
set1 = {1,2,3,4}
set2 = {2,3,5,6}
set1.difference_update(set2)
print(set1)
#union
set1 = {1,2,3,4}
set2 = {2,3,5,6}
set3 = set1.union(set2)
print(set3)
#len
print(len(set3))
#遍历
for element in set3:
    print(element)
````

- 不允许数据重复存在
- 可以修改

## 字典
### 字典的定义
- 可以通过key找到对应的value
- 使用{}，存储的元素是一个个键值对
```python
# 定义字典字面量
{key: value, key:value,...}
# 定义字典变量
my_dict = {key: value, key:value,...}
# 定义空字典
my_dict = {}
my_dict = dict()
```
- 字典的key不可重复，后来的key值所对应的value会覆盖之前的
- 字典只能通过key值来获取对应的value，不可使用下标索引
```python
stu_score = {"jack": 99,"mike": 88,"amy": 67}
print(stu_score["jack"])    #99
print(stu_score["mike"])    #88
print(stu_score["amy"])     #67
```
- 字典中的key和value可以为任意值，但key不可为字典，所以字典可以嵌套

````python
#字典的嵌套
stu_score = {
    "A1":{
        "语文": 77,
        "数学": 66,
        "英语": 33
    },
    "A2":{
        "语文": 88,
        "数学": 86,
        "英语": 55
    },
    "A3":{
        "语文": 99,
        "数学": 96,
        "英语": 66
    }
}
print(stu_score)
print(stu_score["A1"])
````

### 字典的常用操作
- 新增元素：dict[key] = value,key不存在
- 更新元素：dict[key] = value,key存在
- 删除元素：dict.pop(key)
- 清空字典：dict.clear()
- 获取所有的key: dict.keys()
- 遍历：只能用for
- len:字典中的元素个数

````python
my_dict = {"jack": 99,"mike": 88,"amy": 67}
#新增
my_dict["homen"] = 80
print(my_dict)
# 更新
my_dict["jack"] = 85
print(my_dict)
# 删除
my_dict.pop("homen")
print(my_dict)
# 清空
my_dict.clear()
print(my_dict)
# 获取所有的key
my_dict = {"jack": 99,"mike": 88,"amy": 67}
keys = my_dict.keys()
print(keys)
# 遍历
# 方法一：通过获取到全部的key来遍历
for key in keys:
    print(key)
    print(my_dict[key])
# 方法二：直接对自带你循环，每次都得到key和value
for key in my_dict:
    print(key)
    print(my_dict[key])
# 元素个数
print(len(my_dict))
````

### 字典的特点
- 可以容纳多个数据
- 可以容纳不同类型的数据
- 每一份数据都是键值对
- 可以通过key获取value，key不重复
- 不支持下标索引
- 可以修改
- 支持for循环，不支持while循环

## 数据容器的通用操作
- 遍历操作
- 统计操作：
    - len(容器)：统计容器元素的个数
    - max(容器)：统计容器的最大元素
    - min(容器)：统计容器最小元素
- 类型转换：
    - list(容器)：转换为列表
    - tuple(容器)：转换为元组
    - str(容器)：转换为字符串
    - set(容器)：转换为集合
    - 字典转换列表，元组，集合后会舍弃value
- 排序：
    - sorted(容器, [reverse = True])为反向排序
    - sorted(容器)为正向排序
    - 排序结果将变成列表，且字典会舍弃value

````python
my_list = [1,2,3,4,5]
my_tuple = (1,2,3,4,5)
my_str = 'abcdefg'
my_set = {1,2,3,4,5}
my_dict = {'key1': 1,'key2': 2,'key3': 3,'key4': 4,'key5': 5}
## len
print(len(my_list))
print(len(my_tuple))
print(len(my_str))
print(len(my_set))
print(len(my_dict))
# max
print(max(my_list))
print(max(my_tuple))
print(max(my_str))
print(max(my_set))
print(max(my_dict))
# min
print(min(my_list))
print(min(my_tuple))
print(min(my_str))
print(min(my_set))
print(min(my_dict))
````
