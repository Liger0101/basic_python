# 装饰器和生成器
## 装饰器
- 在不修改原有代码的情况下，对功能进行修改
- 一般语法
```python
def 函数名(func):
    def 内置函数():
        func()
    return 内置函数
```

````python
def outer(func):
    def inner():
        # 放此处则在执行原函数之前触发
        print("准备开始")
        func()
        # 放此处则在执行原函数之后触发
        print('执行结束')
    return inner    #装饰器函数

@outer
def send_wechat():
    print('微信')

@outer
def send_email():
    print('邮件')

@outer
def send_sms():
    print('短信')

if __name__ == '__main__':
    send_wechat()
    send_email()
    send_sms()
````

- 原理解释

````python
def outer(func):
    def inner():
        # 放此处则在执行原函数之前触发
        print("准备开始")
        func()  # 原函数
        # 放此处则在执行原函数之后触发
        print('执行结束')
    return inner    #装饰器函数

@outer  # 执行outer(send_wechat)，随后返回赋值send_wechat = outer(send_wechat)
def send_wechat():
    print('微信')

if __name__ == '__main__':
    send_wechat()
````

- 当函数有参数时，需要给内置函数的括号内设定参数，并在原函数内也设定同样的参数
- 如果有多个参数时，内部可以写*args,**kwargs，这样无论多少参数都可以输入，参考可变参数

````python
def outer(func):
    def inner(*args,**kwargs):   # 参数先传给内置函数
        # 放此处则在执行原函数之前触发
        print("准备开始")
        func(*args,**kwargs)  # 原函数 #然后再传给原函数
        # 放此处则在执行原函数之后触发
        print('执行结束')
    return inner    #装饰器函数

@outer  # 执行outer(send_wechat)，随后返回赋值send_wechat = outer(send_wechat)
def send_wechat(body):
    print('微信',body)

if __name__ == '__main__':
    send_wechat('hello')
````

- 正常如果函数有返回值，可用变量接受

````python
def send_wechat(body):
    print('微信',body)
    return 100

if __name__ == '__main__':
    res = send_wechat('hello')
    print(res)
````

- 如果直接使用上述的装饰器，则会没有返回值

````python
def outer(func):
    def inner(*args,**kwargs):   # 参数先传给内置函数
        # 放此处则在执行原函数之前触发
        # print("准备开始")
        func(*args,**kwargs)  # 原函数 #然后再传给原函数 #此处没有接收返回值
        # 放此处则在执行原函数之后触发
        # print('执行结束')
    return inner    #装饰器函数

@outer
def send_wechat(body):
    print('微信',body)
    return 100

if __name__ == '__main__':
    res = send_wechat('hello')
    print(res)
````

- 所以需要一个变量接受返回值

````python
def outer(func):
    def inner(*args,**kwargs):   # 参数先传给内置函数
        # 放此处则在执行原函数之前触发
        # print("准备开始")
        value = func(*args,**kwargs)  # 原函数 #然后再传给原函数 #此处用value接收返回值
        # 放此处则在执行原函数之后触发
        # print('执行结束')
        return value
    return inner    #装饰器函数

@outer
def send_wechat(body):
    print('微信',body)
    return 100

if __name__ == '__main__':
    res = send_wechat('hello')
    print(res)
````

## 生成器
- 当一个函数体内包含 yield 时，调用该函数不会立即执行函数体，而是返回一个生成器对象。
- 执行流程：只有当你对该生成器调用 next() 或进行 for 循环迭代时，函数才会开始执行。
- 暂停与恢复：执行到 yield 行时，函数会“冻结”当前所有局部变量的状态，返回 yield 后面的值。下次迭代时，函数从冻结处恢复，继续往下执行。

````python
def my_gen():
    print("开始")
    yield 1
    print("继续")
    yield 2
    print("结束")

g = my_gen()
print(next(g))  # 输出: 开始 \n 1
print(next(g))  # 输出: 继续 \n 2
print(next(g))  # 输出: 结束，然后抛出 StopIteration 异常
````
