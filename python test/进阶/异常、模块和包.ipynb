# 异常、模块、包
## 捕获异常
### 捕获常规异常
- 对可能出现的bug进行处理即为捕获异常
- 一般有两种处理bug的情况
    - 整个程序因为该bug停止
    - 对bug进行提醒，程序继续运行
- 捕获异常的作用在于：提前假设某处有异常，提前做好准备，当真的出现异常时，可以有后续手段
- try: 可能发生错误的代码
- except：代码的处理方法

````python
try:
    f = open("D:/pycharm.txt","r",encoding="utf-8")
except:
    print("文件异常")
    f = open("D:/pycharm.txt","w",encoding="utf-8")
````

### 捕获指定异常
```python
try:
    print(name)
except NameError as e:
    print('name变量未定义错误')
```
- 如果尝试执行的代码异常类型与捕获的不同，则无法捕获异常
- 一般下try下方只放一行尝试执行的代码

### 捕获多个异常
```python
try:
    print(1/0)
except(NameError,ZeroDivisionError) as e:
    print("出现异常")
```
- 利用该写法可得到多个异常捕获
- 还可以捕获所有异常
    - 使用try：except:
    - 使用try: except Exception as e:


### 异常else和finally
- else表示如果没有异常时执行的代码
```python
try:
    print(1/0)
except Exception as e:
    print("出现异常")
else:
    print('没有异常')
```
- finally表示无论是否有异常都要执行的代码，比如关闭文件
```python
try:
    f = open("D:/w.text",'r')
except Exception as e:
    print("出现异常")
    f = open("D:/w.text",'w')
else:
    print('没有异常')
finally:
    f.close()
```

## 异常的传递
- 例如在函数调用中，由于重复调用，会导致异常传递

````python
def fuc1():
    print("func1开始执行")
    num = 1 / 0
    print("结束")

def fuc2():
    fuc1()

def main():
    try:
        fuc2()
    except Exception as e:
        print('出现异常了')
        print(e)

main()
````

## 模块
### 模块的导入
- 就是一个python文件，模块能定义函数，类，变量，相当一个工具包
- 导入语法
    - import 模块名
    - from 模块名 import 类，变量，方法
    - from 模块名 import *
    - import 模块名 as 别名
    - from 模块名 import 功能名 as 别名
    - 其中只有import是必须要写的，其余均可省略
- 使用import导入，后面调用则需要写：模块名.功能名()，调用了所有功能
- 使用from导入，from 模块名 import 功能名，后面调用时只需要：功能名(),调用了指定功能
- 使用from导入，from 模块名 import *，后面调用时只需要：功能名(),调用了所有功能
- 使用别名时，可以将上述的功能名或者模块名换为别名

### 自定义模块
- 通过创建一个python文件作为模块，在另一个文件中，使用import即可调用该模块

````python
from model import *
test(1,2)
````

- 当导入多个模块时，模块内部有同名的功能时，调用该功能时，调用的时后面导入的模块的功能
- 在开发中一般会在模块中加入测试信息，比如test(1,1)，此时调用会使用模块内的测试,可在模块中加入__main__解决,从而实现测试也不会在导入时运行测试
- 如果一个模块文件中有__all__变量，则import *时，则只能导入该列表中的函数

## python包
### 自定义包
- 可以看作一个文件夹，文件夹内包括各个模块以及一个__init__.py文件，本质上也是个模块
- 导入规则：import package.model
- 同模块，可用__all__定义*，在init文件中定义
