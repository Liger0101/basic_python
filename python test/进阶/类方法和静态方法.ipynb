# 类方法和静态方法
## 类方法
- 在面向对象的学习中，类内部的方法常用先定义一个对象然后通过对象使用，即实例方法
- 若是直接用类来写会发生报错，因为self变量没有传参

```python
class myclass:
    def funl(self,s):
        print("<<<",s,">>>")

c = myclass()
c.funl("hello")
try:
    myclass.funl("world")   #报错
except Exception as e:
    myclass.funl(c, "world")    #将实例传入self中，此时就不报错了
```

- 类中的方法分为三类：实例方法，类方法以及静态方法
- 类方法的构建，需要通过装饰器构建，即@classmethod

```python
class myclass:
    def funl(self,s):
        print("<<<",s,">>>")

    @classmethod
    def fun2(cls,s):
        print('111',s,'111')

myclass.fun2('s')   # 和实例方法不同，没有报错
```

- 与实例方法不同，类方法用实例调用也可以使用

## 静态方法
- 同样需要装饰器，为@staticmethod
- 静态方法与普通函数定义一样，不需要传入self和cls，因而无法访问类属性和对象属性

```python
class math:
    pass

def myadd(x, y):    #此时为全局函数
    return x + y

m = math()

result = myadd(100,200)
print(result)
```

```python
#接下来将该函数放到类内，将其定义为静态方法
class math:
    @staticmethod
    def myadd(x, y):    #此时为静态方法
        return x + y

m = math()

result = m.myadd(100,200)   # 可以用实例调用
print(result)
result = math.myadd(200,200)    #也可以用类调用
print(result)
```
